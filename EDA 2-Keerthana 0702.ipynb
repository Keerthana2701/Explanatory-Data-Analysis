### working of zip and  json.loads

```python
a = ("John", "Charles", "Mike")
b = ("Jenny", "Christy", "Monica", "Vicky")

x = zip(a, b)

print(tuple(x))
```

```python
import json

# some JSON:
x =  '{ "name":"John", "age":30, "city":"New York"}'

# parse x:
y = json.loads(x)

# the result is a Python dictionary:
print(y["age"])
```

```python
def parse_col_json(column, key):
    """
    Args:
        column: string
            name of the column to be processed.
        key: string
            name of the dictionary key which needs to be extracted
    """
    for index,i in zip(moviesds.index,moviesds[column].apply(json.loads)):
        list1=[]
        for j in range(len(i)):
            list1.append((i[j][key]))# the key 'name' contains the name of the genre
        moviesds.loc[index,column]=str(list1)
    
parse_col_json('genres', 'name')
parse_col_json('spoken_languages', 'name')
parse_col_json('cast', 'name')
parse_col_json('production_countries', 'name')
```

### we can write a generic function using this as min-max function so that any column can use this like budget


```python
def min_max_in(col):    
    
    idof1 = moviesds[col].idxmax()   ## Note: every line shoud be followed by tab
    idof2 = moviesds[col].idxmin()

    high=pd.DataFrame(moviesds.loc[idof1])
    low=pd.DataFrame(moviesds.loc[idof2])
    result=pd.concat([high,low],axis=1)
    return result

min_max_in('budget')
```
